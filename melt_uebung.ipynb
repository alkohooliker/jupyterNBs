````python
import pandas as pd
````

```Dataframe``` | *obligatorisch* | Es ist der ```DataFrame```, den wir in das lange Format wechseln wollen.

```id_vars``` | *optional* | Es kann ein ```Tupel```, eine ```Liste``` oder ein N-dimensionales ```Array``` sein. Es ist die Spalte, die für Identifizierungsvariablen verwendet wird. Sie können mehr als eine Bezeichnerspalte auswählen.

```value_vars``` | *optional* | Es kann ein ```Tupel```, eine ```Liste``` oder ein N-dimensionales ```Array``` sein. Standardmäßig sind die Spalten, die nicht als Bezeichnungsvariablen angegeben sind, Wertvariablen. Sie können diese auch auswählen.

```var_name``` | *optional* | Es ist eine Variable vom Typ ```scalar```. Es ist der Name der Bezeichnerspalte. Standardmäßig ist sie ```variabel```.

```value_name``` | *optional* | Es ist eine Variable vom Typ ```scalar```. Es ist der Name der Bezeichnerspalte. Standardmäßig ist sie ```variabel```.


```col_level``` | *optional* | Es ist eine Ganzzahl oder eine Zeichenkette. Im Falle von Multi-Index-Spalten können wir diesen Parameter verwenden, um unseren ```DataFrame``` zu transformieren.

````python
# Pandas Dataframe erstellen

df = pd.DataFrame({'Person': ['Alan','Berta','Charlie', 'Danielle']
                    ,'House': ['A', 'B', 'A', 'D']
                    ,'Age': [32, 46, 35, 28 ]
                    , 'Books': [100, 30, 20, 40]
                    ,'Movies': [10, 20, 80, 60]
})

````

````python
data_wide = df
data_wide
````

````python
# melt-Funktion wird auf Dataframe, ohne weiter Parameter
data_wide.melt()
````

Verwendete Parameter:

id_vars: Person, Haus

In diesem Beispiel schmelzen wir unseren data_wide DataFrame, indem wir nur den id_vars Parameter angeben . Die in id_vars(Person, Haus) verwendeten Spalten werden als Bezeichnungsvariablen bezeichnet, und diese Spalten werden nicht geschmolzen. Da wir keine anderen Parameter angeben, werden alle anderen Spalten (Alter, Bücher, Filme) zu einer einzigen Spalte geschmolzen.

In der Ausgabe haben wir zwei neue Spalten: Variable und Wert. In der variablen Spalte werden die Namen der geschmolzenen Spalten und in der Wertespalte die tatsächlichen Werte der geschmolzenen Spalten gespeichert. Wir können die Namen der Ausgabespalten mit var_nameund value_nameParametern ändern .

Das Entscheidende dabei ist, dass durch Nichtangabe der zu schmelzenden Spalten alle verbleibenden Spalten nach Angabe der Spaltennamen in geschmolzen werden id_vars.

````python
data_wide.melt(id_vars=['Person', 'House'])
````

## D. Beispiel - 3: ##

Verwendete Parameter:

*id_vars*: Person, Haus

value_vars: Alter, Bücher, Filme

In diesem Beispiel geben wir ```id_vars``` an, dass die Spaltennamen angegeben werden sollen, die nicht geschmolzen werden sollen (Person, Haus). Ansonsten verwenden wir den ```value_vars``` Parameter, um die Spaltennamen anzugeben, die wir schmelzen möchten (Alter, Bücher, Filme).

````python
data_wide.melt(
    id_vars=['Person', 'House']
    , value_vars=['Age', 'Books', 'Movies']
    )
````

## F. Beispiel - 5: ##
Verwendete Parameter:

```id_vars```: Person

```value_vars```: Bücher, Filme

```var_name```: Medium

Wertname: Numerisch

Das folgende Beispiel zeigt, wie wir die Parameter ```var_name``` und ```value_name``` verwenden können , um den Namen der Ausgabespalten zu ändern.

````python
data_wide.melt(
    id_vars=['Person']
    , value_vars=['Movies', 'Books']
    , var_name='Info'
    , value_name= 'Numerical'
    )
````

Im folgenden Beispiel werden ausgewählte Spalten für ```id_vars``` und ```value_vars``` Parameter verwendet. Abgesehen davon ändern wir auch die Spaltennamen des ausgegebenen DataFrame mithilfe von ```var_name``` und ```value_name``` -Parametern.

````python
data_wide.melt(
    id_vars=['Person']
    , value_vars=['Books', 'Movies']
    , var_name='Medium'
    , value_name= 'Anzahl'
    )
````

## H. Beispiel - 7:##
Verwendete Parameter:

```id_vars```: Person, Haus

```value_vars```: Alter, Bücher, Filme

```var_name```: Info

```Wertname```: Numerisch

```Ignore_index```: False

Das folgende Beispiel zeigt die Verwendung des ```ignore_index``` Parameters zum Ignorieren oder Beibehalten des ursprünglichen Index. Wenn wir setzen ```ignore_index``` auf ```False```, dann wird der ursprüngliche Index beibehalten. Wenn wir setzen ```ignore_index``` auf ```True```, wird es den ursprünglichen Index und assign kontinuierliche Werte für den Index ignorieren.

````python
data_wide.melt(
    id_vars=['Person', 'House']
    , value_vars=['Age', 'Books', 'Movies']
    , var_name='Info'
    , value_name= 'Numerical'
    , ignore_index= True
    )
````

````python
data_wide.melt(
    id_vars=['Person', 'House']
    , value_vars=[ 'Books','Age', 'Movies']
    , var_name='Info'
    , value_name= 'Numerical'
    , ignore_index= False
    )
````

## I. Erstellen eines DataFrame mit mehreren Indizes:##
In einigen Fällen gibt es mehrere Namen für die Spalten. In solchen Fällen können wir den ```col_level``` Parameter verwenden, um anzugeben, welche Ebene beim Schmelzen des DataFrame verwendet werden soll.

````python
data_wide.columns = [['Person', 'House', 'Age', 'Books','Movies']
                     ,['Name', 'Flat', 'Old', 'Text', 'Video']
                     ]

data_wide
````

## J. Beispiel - 8:##
Verwendete Parameter:

```id_vars```: Person, Haus

```value_vars```: Alter, Bücher, Filme

```var_name```: Info

```Wertname```: Numerisch

```col_level```: 0

Das folgende Beispiel zeigt, wie wir den ```col_level``` Parameter verwenden können, um den DataFrame mithilfe des 0. Index der Spaltenebenen zu schmelzen.

````python
data_wide.melt(
    id_vars= ['Person', 'House']
    , value_vars= ['Age', 'Books', 'Movies']
    , var_name= 'Info'
    , value_name= 'Numerical'
    , col_level= 0
)
````
